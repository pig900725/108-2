# 3.猜數字

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
```

```python
rang1 = int(input("最小值:"))
rang2 = int(input("最大值:"))
```

```python
num = random.randint(rang1,rang2)
guess = "guess"
```

```python
print("數字遊戲")
i = 0
while guess != num:
    i += 1
    guess = int(input("猜的數字："))

    if guess == num:
        print("答對了")
    elif guess < num:
        print("太小囉")
    else:
        print("太大惹")

print("你總共猜了%d" %i + "次",end = '')
print("真的爛")
```
