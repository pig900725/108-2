```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## 1.載入套件

```python
from tensorflow.keras.datasets import mnist
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train.shape
```

## 2.轉換資料格式 (28,28)變(28,28,1)

```python
x_train = x_train.reshape(60000,28,28,1) / 255
x_test = x_test.reshape(10000,28,28,1) / 255
```

```python
x_train[87].shape
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

## 3.讀入必要函式

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import SGD
```

## 4.打造函數學習機

```python
model = Sequential()
```

### 第一層

```python
model.add(Conv2D(16, (3,3), padding='same',
                input_shape=(28,28,1),
                activation='relu'))
```

### 第二層

```python
model.add(Conv2D(32,(3,3),padding='same',
                activation= 'relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
```

### 第三層

```python
model.add(Conv2D(64,(3,3),padding='same',
                activation= 'relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
```

#### dense 54 ->35

```python
model.add(Flatten())
```

```python
model.add(Dense(35, activation='relu'))
```

```python
model.add(Dense(10, activation='softmax'))
```

## 5.檢視神經網路

```python
model.summary()
```

#### 設定learning_Rate為0.09,batch_size為200,epochs為10

```python
model.compile(loss="mse",optimizer=SGD(lr=0.09),
             metrics=["accuracy"])
```

```python
model.fit(x_train,y_train,batch_size=100,epochs=20)
```

## 5.建立另一個函數學習機

```python
model2= Sequential()
```

### 第一層

```python
model2.add(Conv2D(32,(3,3),padding='same',
                input_shape=(28,28,1),
                activation= 'relu'))
model2.add(MaxPooling2D(pool_size=(2,2)))
```

### 第二層

```python
model2.add(Conv2D(64,(3,3),padding='same',
                activation= 'relu'))
model2.add(MaxPooling2D(pool_size=(2,2)))
```

### 第三層

```python
model2.add(Conv2D(128,(3,3),padding='same',
                activation= 'relu'))
model2.add(MaxPooling2D(pool_size=(2,2)))
```

```python
model2.add(Flatten())
```

```python
model2.add(Dense(70,activation='relu'))
model2.add(Dense(10,activation='softmax'))
```

```python
model2.summary()
```

```python
model2.compile(loss="mse",optimizer=SGD(lr=0.09),
             metrics=["accuracy"])
```

```python
model2.fit(x_train,y_train,batch_size=300,epochs=20)
```

## 6.預測

```python
result = model2.predict_classes(x_test)
```

```python
def my_predict(n):
    print('我的預測是', result[n])
    X = x_test[n].reshape(28,28)
    plt.imshow(X)
```

```python
from ipywidgets import interact_manual
```

```python
interact_manual(my_predict, n=(0, 9999))
```

```python
score = model2.evaluate(x_test, y_test)
```
